```python
from posixpath import sep
import pandas as pd

file_path = '/content/drive/MyDrive/SMSSpamCollection.txt'

messages = pd.read_csv(file_path, sep='\t', names=["labels", "message"])

messages
```

```python
messages.shape, messages['message'].loc[451]
```

```python
# data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
X.shape, X[1]
```

```python
y = pd.get_dummies(messages['labels'])
y = y.iloc[:, 1].values
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train, y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
# prediction
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```

## Word2Vec Implementation on SMS spam collection dataset

- Gensim has word2vec in it, there are 2 ways of using word2vec
>1.   Pre-Trained Model - eg: Google News 300, etc
>2.   Train your own model

- word2vec can be done using skipgrams/CBOW.

But, if I use google news 300, then for input "I want to eat pizza" would have 5X300 shape as each word would be converted into 1X300 dimension for the google model.

To fix this problem, I can use avgWord2Vec.

avgWord2Vec says, each input feature in the 1X300 in all words will be added and averaged to get just one 1X300 as input for the pre-trained model.

eg: Image 1 in word2vec section.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
corpus = []

for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
  review = review.lower()
  review = review.split()

  review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt')
words = []
for sent in corpus:
  sent_token=sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))

words
```

- based on unique words, I will now create my model from scratch

```python
import gensim
model = gensim.models.Word2Vec(words, window=5, min_count=2) # window is used to understand the context before and after, also used in CBOW
model.wv.index_to_key
```

```python
model.corpus_count, model.epochs
```

```python
model.wv.similar_by_word('hope')
```

```python
model.wv['kid'].shape
```

```python
import numpy as np
def avg_word2vec(doc):
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis=0)
```
